```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

```python
data = pd.read_csv('./Churn_Modelling.csv')
data.head()
```

```python
df = data
```

```python
##Preprocesse the data
## Drop the columns which are unique for all users like IDs
df = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
```

```python
## Encode the categorical data
label_encoder_gender = LabelEncoder()
df['Gender'] = label_encoder_gender.fit_transform(df['Gender'])
df
```

```python
## One hot encoding
## One way to do one hot encoding
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

columnTransformer = ColumnTransformer([('encoder', OneHotEncoder(), [1])], remainder='passthrough')
df = columnTransformer.fit_transform(df)
df = pd.DataFrame(df)
df
```

```python
df.head()
```

```python
onehot_encoder_geo = OneHotEncoder()
geo_encoder = onehot_encoder_geo.fit_transform(data[['Geography']])
geo_encoder
```

```python
onehot_encoder_geo.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df=pd.DataFrame(geo_encoder.toarray(), columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
## Concatenate the dataframe
df = pd.concat([df.drop('Geography', axis=1), geo_encoded_df], axis=1)
```

```python
df
```

```python
## save the encoders and scalers
with open('artifacts/label_encoder_gender.pkl', 'wb') as file:
    pickle.dump(label_encoder_gender, file)

with open('artifacts/onehot_encoder_geo.pkl', 'wb') as file:
    pickle.dump(onehot_encoder_geo, file)
```

```python
X = df.drop('Exited', axis=1)
y = df['Exited']
```

```python
## Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
## Save the scaler
with open('artifacts/scaler.pkl', 'wb') as file:
    pickle.dump(scaler, file)
```

```python
#Implement ANN
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
model=Sequential([
    Dense(64,activation='relu', input_shape=(X_train.shape[1],)), #Hidden Layer 1 connected to input layer
    Dense(32,activation='relu'), #Hidden Layer 2
    Dense(1,activation='sigmoid') #Output Layer
])
```

```python
model.summary()
```

```python
X_train.shape
```

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
##setup the tensorboard 

log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
## SetUp early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=40, restore_best_weights=True)
```

```python
## Training the model
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=100,
    
    callbacks=[early_stopping, tensorflow_callback]
)
```

```python
history_df = pd.DataFrame(history.history)
history_df['accuracy'].mean()
```

```python
model.save('artifacts/ann_model.h5')
```
